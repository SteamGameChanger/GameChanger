```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
import numpy as np
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
import re
```

```python
df = pd.read_csv("./drive/MyDrive/BigData/combined_data_for_similar_game_finder_review_and_description.csv",encoding='cp949')
```

```python
corpus = []
is_na=[]
stop_words = set(stopwords.words('english')) 
for i in range(len(df['Currently popular'])):
  if pd.isna(df['Currently popular'][i]):
    is_na.append(1)
    continue
  else:
    example=df['Currently popular'][i]
    result = []
    word_tokens = word_tokenize(example)
    
    for text in word_tokens:
      text = re.sub('[^a-zA-Z0-9]','',text).strip()
      if text not in stop_words:
        result.append(text)
    result=' '.join(result).split()
    corpus.append(' '.join(result))
    is_na.append(0)
```

```python
tfidfv = TfidfVectorizer().fit(corpus)
tmp=tfidfv.transform(corpus).toarray()
```

```python
for i in range(len(tmp)-1,-1,-1):
  if(is_na[i]==1):
    continue
  else:
    df['Currently popular'][i]=tmp[-1]
    np.delete(tmp,-1,0)
```

```python
tfidfv2 = TfidfVectorizer().fit(corpus2)
tmp2=tfidfv2.transform(corpus2).toarray()
```

```python
for i in range(len(tmp2)-1,-1,-1):
  if(is_na2[i]==1):
    continue
  else:
    df['Game Description'][i]=tmp2[-1]
    np.delete(tmp2,-1,0)
```
