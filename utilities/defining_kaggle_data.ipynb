```python
import pandas as pd
```

```python
steam_data = pd.read_csv('filtered_games.csv')
```

```python
steam_data
```

```python
# 데이터 타입 확인
steam_data.dtypes
```

```python
# Estimated owners 열의 스트링 타입 데이터 'k - n'을 [int(k),int(n)] 으로 형식 전환

for i in range(0,len(steam_data)):
    val = steam_data['Estimated owners'][i].split('-')
    try:
        steam_data['Estimated owners'][i] = [int(val[0]), int(val[1])]
    except:
        print(i, ':',val)
    
steam_data
```

```python
# 형식 바뀐 것 확인
steam_data.dtypes
```
