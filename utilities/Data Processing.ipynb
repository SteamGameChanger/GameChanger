```python
# df = pd.read_csv('/content/drive/MyDrive/Big Data Processing/games.csv')
df = pd.read_csv('/content/drive/MyDrive/23-1_online/빅데이터 처리/팀플/Big Data Processing/games.csv')
```

```python
df = df.drop(df.loc[df['Name'].isna()].index)
```

```python
df['Release date cleaned'] = pd.to_datetime(df['Release date'], errors='coerce')
```

```python
df = df.drop(df.loc[df['About the game'].isna()].index)
```

```python
keywords = ['beta', 'alpha', 'test', 'sdk', 'demo', 'server', 'editor']
pattern = '|'.join(keywords)

# Case insensitive filtering
df_filtered = df[~df['Name'].str.contains(pattern, case=False, regex=True)]
```

```python
df_filtered = df_filtered[df_filtered['Price'] != 0]
```

```python
# df_filtered.to_csv('cleaned_games.csv', index=False)
df_filtered.to_csv('/content/drive/MyDrive/23-1_online/빅데이터 처리/팀플/Big Data Processing/cleaned_games.csv', index=False)
```
