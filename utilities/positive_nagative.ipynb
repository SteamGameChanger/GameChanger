```python
import pandas as pd
```

```python
df = pd.read_csv("./drive/MyDrive/BigData/combined_data_for_similar_game_finder.csv",encoding='cp949')
```

```python
print(df['Overall Reviews'])
```

```python
for index, value in df['Overall Reviews'].items():
    if 'positive' in str(value).lower():
        df.loc[index, 'Overall Reviews'] = 0
    elif 'mixed' in str(value).lower():
        df.loc[index, 'Overall Reviews'] = 1
    elif 'negative' in str(value).lower():
        df.loc[index, 'Overall Reviews'] = 2
    elif 'user reviews' in str(value).lower():
        df.loc[index, 'Overall Reviews'] = 3
    elif pd.isna(value):
        df.loc[index, 'Overall Reviews'] = 3
```

```python
print(df['Overall Reviews'])
```

```python
df.to_csv("./drive/MyDrive/combined_data_for_similar_game_finder_2.csv", index=False)
```
